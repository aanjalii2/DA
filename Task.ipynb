Outlier Detection and Treatment

```python
# Function to remove outliers based on IQR
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
```

```python
# Apply outlier removal to selected important features
important_cols = ['Order', 'PID', 'MS SubClass', 'MS Zoning', 'Lot Frontage', 'Lot Area', 'Street', 'Lot Shape', 'Land Contour', 'Utilities', 'Lot Config', 'Land Slope', 'Neighborhood', 'Condition 1', 'Condition 2', 'Bldg Type', 'House Style', 'Overall Qual', 'Overall Cond', 'Year Built', 'Year Remod/Add', 'Roof Style', 'Roof Matl', 'Exterior 1st', 'Exterior 2nd', 'SalePrice']
```

```python

# Now proceed with outlier removal
important_cols =['SalePrice', 'Gr Liv Area', 'Lot Area', 'Order']
```

```python
# Before removing outliers - Boxplots
for col in important_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[col])
    plt.title(f"Before Outlier Removal - {col}")
    plt.show()
```

```python
# Remove outliers one column at a time
for col in important_cols:
    df = remove_outliers_iqr(df, col)
```

```python
for col in important_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[col])
    plt.title(f"After Outlier Removal - {col}")
    plt.show()
```

Normalize or scale numerical features


```python
# Importing StandardScaler
from sklearn.preprocessing import StandardScaler
```

```python
# Selecting only the numerical features
numerical_features = df.select_dtypes(include=[np.number])
```

```python

# Initialize StandardScaler
scaler = StandardScaler()
```

```python
# Fit and transform the numerical features
scaled_numerical = scaler.fit_transform(numerical_features)
```

```python
# Convert scaled data back into a DataFrame
scaled_numerical_df = pd.DataFrame(scaled_numerical, columns=numerical_features.columns)
```

```python
# Replace the numerical columns in original dataframe with scaled values
# (If you want to keep both original and scaled, make a copy before replacing)
for col in numerical_features.columns:
    df[col] = scaled_numerical_df[col]
```
